```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS
from pyspark.sql import Row

from answers.answer import init_spark
spark = init_spark() # spark session (contains context and many other)
sc = spark.sparkContext # spark context
```

```python
sc.version
```

```python
lines = spark.read.text("data/sample_movielens_ratings.txt").rdd
```

```python
# split 0::2::3::123125 data format into list to indicate user, movie, rating and timestamp
lines = lines.map(lambda row: row.value.split("::"))
# give a 'column' name for each value. re-convert into Row() objects
lines = lines.map(lambda row: Row(userid=int(row[0]), movieid=int(row[1]), rating=float(row[2]), timestamp=int(row[3])))
# use this new RDD to create dataframe
ratings = spark.createDataFrame(lines)
# separate into training and test data
(training, test) = ratings.randomSplit([0.8, 0.2])
```

```python
als = ALS(maxIter=5, regParam=0.01, userCol="userid", itemCol="movieid", ratingCol="rating", coldStartStrategy="drop" )
model = als.fit(training)
```
